```python
from pandas_datareader import data
import datetime 
from bokeh.plotting import figure, show, output_file

start = datetime.datetime(2021,12,1)
end = datetime.datetime(2022,3,11)

df=data.DataReader(name="AAPL",data_source="yahoo",start=start, end=end)
```

```python
df
```

```python
p=figure(x_axis_type='datetime',width=1000,height=300,sizing_mode="scale_width")
p.title="Candle Stick Chart "
p.grid.grid_line_alpha = 0.3
```

```python
def inc_dec(o,c):
    if o>c:
        value="Decrease"
    elif o<c:
        value="Increase"
    else:
        value="equal"
    return value

df["Status"]=[inc_dec(o,c) for o,c in zip(df.Open,df.Close)]
```

```python
hour_sec=12*3600*1000
p.segment(df.index,df.High,df.index,df.Low)
p.rect(df.index[df.Status=="Increase"],(df.Close+df.Open)/2, hour_sec, abs(df.Close-df.Open),fill_color="#00FA9A",line_color="black")
p.rect(df.index[df.Status=="Decrease"],(df.Close+df.Open)/2, hour_sec, abs(df.Close-df.Open),fill_color="#FF3333",line_color="black")

show(p)
```
